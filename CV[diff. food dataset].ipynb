```python
import glob
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from joblib import dump
```

```python
# Pre-proccessing the data set

Data_List=[]
Label_list=[]

for i,address in enumerate(glob.glob(f"{folder_path}\\*\\*")):
    img = cv2.imread(address)
    img = cv2.resize(img,(64,64))
    img = img / 255.0
    img = img.flatten()
    Data_List.append(img)
    
    p = address.split("\\")[-1].split(".")[0]
    Label_list.append(p)
Data_List = np.array(Data_List)
```

```python
print(Data_List[50])
print(Label_list[50])
```

```python
x_train,x_test,y_train,y_test = train_test_split(Data_List,Label_list,test_size=0.2)

classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(x_train,y_train)

dump(classifier,"detectFood.z")

accuracy_rate = classifier.score(x_test,y_test)

print(accuracy_rate*100)
```
